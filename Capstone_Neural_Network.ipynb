```python
def import_image_data(path,piece_label,color_label):
    '''
    Function to read in image data and store to arrays
    '''
    X_data = []
    y_pdata = []
    y_cdata = []
    fname_data = []
    
    #glob string for looping through all files
    glob_string = path + '/*.jpg'
    for img in glob.glob(glob_string):
        #Read in image and resize to 375x500
        img_data = cv2.imread(img,0)
        X_data.append(img_data)
        y_pdata.append(piece_label)
        y_cdata.append(color_label)
        fname_data.append(img)
    return X_data, y_pdata, y_cdata, fname_data
```

```python
#make labels nparrays
y_piece_train = np.array(y_piece_train)
y_piece_test = np.array(y_piece_test)
y_piece_valid = np.array(y_piece_valid)

#make labels nparrays
y_color_train = np.array(y_color_train)
y_color_test = np.array(y_color_test)
y_color_valid = np.array(y_color_valid)
```

```python
from keras.utils import np_utils
# one-hot encode the labels
num_classes = len(np.unique(y_piece_train))
print(num_classes)

y_piece_train_cat = keras.utils.to_categorical(y_piece_train, num_classes)
y_piece_test_cat = keras.utils.to_categorical(y_piece_test, num_classes)
y_piece_valid_cat = keras.utils.to_categorical(y_piece_valid, num_classes)


# one-hot encode the labels
num_classes = len(np.unique(y_color_train))
print(num_classes)

y_color_train_cat = keras.utils.to_categorical(y_color_train, num_classes)
y_color_test_cat = keras.utils.to_categorical(y_color_test, num_classes)
y_color_valid_cat = keras.utils.to_categorical(y_color_valid, num_classes)
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale = 1/255.0,
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True)

datagen.fit(X_train_aug)
```

```python
#Define NN architecture

from keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D, BatchNormalization
from keras.layers import Dropout, Flatten, Dense
from keras.models import Sequential

model = Sequential()
model.add(Conv2D(filters=16, kernel_size=5, padding='same', activation='relu', 
                        input_shape=(135, 135, 1)))
model.add(MaxPooling2D(pool_size=2))
model.add(BatchNormalization())
model.add(Conv2D(filters=32, kernel_size=3, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(BatchNormalization())
model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(BatchNormalization())
model.add(Conv2D(filters=128, kernel_size=2, padding='same', activation='relu'))
model.add(Conv2D(filters=128, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(BatchNormalization())
model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(BatchNormalization())
model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(GlobalAveragePooling2D())
model.add(Dense(2048, activation='relu'))
model.add(Dense(2048, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(7, activation='softmax'))


model.summary()
```

```python
# compile the model
rms_opt = keras.optimizers.RMSprop(lr=0.0001, rho=0.9, epsilon=None, decay=0.0)
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', 
                  metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint   

# train the model
checkpointer = ModelCheckpoint(filepath='model.weights.best.hdf5', verbose=1, 
                               save_best_only=True)
hist = model.fit(X_train_scaled_2, y_piece_train_cat, batch_size=64, epochs=500,
          validation_data=(X_valid_scaled_2, y_piece_valid_cat), callbacks=[checkpointer], 
          verbose=2, shuffle=True)

#hist = model.fit_generator(datagen.flow(X_train_aug,y_train_cat,batch_size=64),epochs=30,
#                           validation_data=(X_valid_scaled_2, y_valid_cat), callbacks=[checkpointer], 
#                           verbose=2, shuffle=True)
```

```python
#Define NN architecture

from keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D, BatchNormalization
from keras.layers import Dropout, Flatten, Dense
from keras.models import Sequential

color_model = Sequential()
color_model.add(Conv2D(filters=16, kernel_size=5, strides=4, padding='same', activation='relu', 
                        input_shape=(135, 135, 1)))
color_model.add(MaxPooling2D(pool_size=2))
color_model.add(BatchNormalization())
color_model.add(Conv2D(filters=32, kernel_size=2, padding='same', activation='relu'))
color_model.add(MaxPooling2D(pool_size=2))
color_model.add(BatchNormalization())
color_model.add(Conv2D(filters=128, kernel_size=2, padding='same', activation='relu'))
color_model.add(Conv2D(filters=128, kernel_size=2, padding='same', activation='relu'))
color_model.add(MaxPooling2D(pool_size=2))
color_model.add(BatchNormalization())
color_model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
color_model.add(MaxPooling2D(pool_size=2))
color_model.add(BatchNormalization())
color_model.add(Dropout(0.3))
color_model.add(GlobalAveragePooling2D())
color_model.add(Dense(4096, activation='relu'))
color_model.add(Dense(4096, activation='relu'))
color_model.add(Dropout(0.4))
color_model.add(Dense(3, activation='softmax'))

color_model.summary()
```
